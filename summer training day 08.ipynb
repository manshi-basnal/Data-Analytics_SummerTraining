```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("line plot example")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid(True)
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 5, 7]
plt.bar(x, y, color='red')
plt.title("Bar chart")
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 5, 7]
plt.barh(x, y, color='red')
plt.title("Bar chart")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 5, 7, 6]
plt.scatter(x, y, color='red')
plt.title("scatter plot")
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue')
plt.title("Histograph")
plt.show()
```

```python
labels = ['python', 'java', 'c++']
sizes = [50, 30, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie chart")
plt.show()
```

```python
data = [3, 6, 8, 8, 8, 5, 4]
plt.boxplot(data)
plt.title("Box plot")
plt.show()
```
