```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
img = mpimg.imread(r'C:\Users\HP\Desktop\Data analytics_Summer training\rgb\rainbow.webp') 
print("Original Shape:", img.shape)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()

bgr_img = img[..., ::-1]

plt.imshow(bgr_img)
plt.title("BGR Image (Reversed RGB)")
plt.axis('off')
plt.show()

grayscale = 0.299 * img[..., 0] + 0.587 * img[..., 1] + 0.114 * img[..., 2]
grayscale = grayscale.astype(np.uint8)
plt.imshow(grayscale, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

r_df = pd.DataFrame(img[..., 0])
g_df = pd.DataFrame(img[..., 1])
b_df = pd.DataFrame(img[..., 2])

print("Red Channel DataFrame:\n", r_df.head())
print("Green Channel DataFrame:\n", g_df.head())
print("Blue Channel DataFrame:\n", b_df.head())

img_m1 = img.copy()
img_m1[..., 0] = 1
plt.imshow(img_m1)
plt.title("Modified Image (Red Max)")
plt.axis('off')
plt.show()

img_m2 = img.copy()
img_m2[..., 1] = 1
plt.imshow(img_m2)
plt.title("Modified Image (green Max)")
plt.axis('off')
plt.show()

img_m3 = img.copy()
img_m3[..., 2] = 1
plt.imshow(img_m3)
plt.title("Modified Image (blue Max)")
plt.axis('off')
plt.show()
```
