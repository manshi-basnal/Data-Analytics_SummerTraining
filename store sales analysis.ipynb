```python
plt.figure(figsize=(10,5))

# Subplot 1: Line plot
plt.subplot(1, 2, 1)
plt.plot(df['date'], df['sales'], color='blue')
plt.title('Sales Over Time')
plt.xticks(rotation=45)

# Subplot 2: Histogram
plt.subplot(1, 2, 2)
plt.hist(df['sales'], bins=10, color='orange', edgecolor='black')
plt.title('Sales Distribution')

plt.tight_layout()
plt.show()
```

```python
holiday_sales = df[df['holiday'] == 1].groupby('store')['sales'].sum()
nonholiday_sales = df[df['holiday'] == 0].groupby('store')['sales'].sum()

# Ensure same order of stores
stores = nonholiday_sales.index

plt.figure(figsize=(8,5))

# Non-holiday sales as base
plt.bar(stores, nonholiday_sales, label='Non-Holiday Sales', color='skyblue')

# Holiday sales stacked on top
plt.bar(stores, holiday_sales, bottom=nonholiday_sales, label='Holiday Sales', color='lightgreen')

# Labels and title
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.title('Stacked Bar Chart: Holiday vs Non-Holiday Sales')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.boxplot(df['sales'])
plt.ylabel('Sales')
plt.title('Box Plot of Sales')
plt.show()
```

```python
df = df.sort_values('date')

x = df['date']
y = df['sales']
plt.figure(figsize=(10,5))
plt.plot(x, y, color='blue', label='Sales')
plt.fill_between(x, y, color='lightblue', alpha=0.5)  # Fill area under sales curve
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Over Time (Fill Between)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby('store')['sales']
mean_sales = grouped.mean()
std_sales = grouped.std()

plt.figure(figsize=(8,5))
plt.bar(mean_sales.index, mean_sales.values, yerr=std_sales.values,
        capsize=5, color='lightgreen', edgecolor='black')

plt.xlabel('Store')
plt.ylabel('Average Sales')
plt.title('Average Sales per Store with Error Bars (Std Dev)')
plt.tight_layout()
plt.show()
```
