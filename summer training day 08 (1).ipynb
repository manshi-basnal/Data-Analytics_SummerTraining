```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("line plot example")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid(True)
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 5, 7]
plt.bar(x, y, color='red')
plt.title("Bar chart")
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 5, 7]
plt.barh(x, y, color='red')
plt.title("Bar chart")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 5, 7, 6]
plt.scatter(x, y, color='red')
plt.title("scatter plot")
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue')
plt.title("Histograph")
plt.show()
```

```python
labels = ['python', 'java', 'c++']
sizes = [50, 30, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie chart")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar([1, 2, 3], [3, 2, 5])
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3]
boys = [20, 35, 30]
girls = [25, 32, 34]

plt.bar(x, boys, label='Boys')
plt.bar(x, girls, bottom=boys, label='Girls')  # girls ki bars boys ke upar

plt.xlabel('Class')
plt.ylabel('Students Count')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
data = [7, 15, 13, 21, 22, 19, 25, 30, 10, 9, 11]

plt.boxplot(data)
plt.title('Box Plot Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 3, 5]
y2 = [0.5, 1, 2, 2, 3]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5)

plt.title('Fill Between Plot')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 4]
error = [0.3, 0.2, 0.4, 0.1]

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5)
plt.title('Error Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
