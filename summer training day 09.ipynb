```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
```

```python
# Sample dataset (like 'tips')
data = {
    'total_bill': [10, 20, 30, 40, 50, 15, 25, 35, 45, 55],
    'tip': [1, 2, 3, 4, 5, 1.5, 2.5, 3.5, 4.5, 5.5],
    'sex': ['Male', 'Female']*5,
    'smoker': ['Yes', 'No']*5,
    'day': ['Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri'],
    'time': ['Lunch', 'Dinner']*5,
    'size': [1, 2, 3, 4, 1, 2, 3, 4, 2, 3]
}

df = pd.DataFrame(data)

# Set style
sns.set(style='whitegrid')

# 1. Barplot
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Barplot")
plt.show()
# 2. Countplot
sns.countplot(x='day', data=df)
plt.title("Countplot")
plt.show()

# 3. Boxplot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Boxplot")
plt.show()

# 4. Violinplot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violinplot")
plt.show()

# 5. Stripplot
sns.stripplot(x='day', y='total_bill', data=df)
plt.title("Stripplot")
plt.show()
# 6. Swarmplot
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title("Swarmplot")
plt.show()

# 7. Pointplot
sns.pointplot(x='day', y='total_bill', data=df)
plt.title("Pointplot")
plt.show()

# 8. Boxenplot
sns.boxenplot(x='day', y='total_bill', data=df)
plt.title("Boxenplot")
plt.show()

# 9. Histplot
sns.histplot(df['total_bill'], kde=True)
plt.title("Histogram with KDE")
plt.show()

# 10. KDE Plot
sns.kdeplot(df['total_bill'], shade=True)
plt.title("KDE Plot")
plt.show()
# 11. Heatmap (correlation)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()

# 12. Pairplot
sns.pairplot(df)
plt.suptitle("Pairplot", y=1.02)
plt.show()

# 13. Jointplot
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
plt.suptitle("Jointplot", y=1.02)
plt.show()
```

```python
# Step 1: Load the image using matplotlib
img = mpimg.imread('C:\Users\HP\Desktop\dataset2\data.webp')  # Replace with your image path

print("Original Shape:", img.shape)  # Shape should be (height, width, 3)

# Step 2: Show the original image
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Step 3: Convert RGB  BGR (just reverse the last axis)
bgr_img = img[..., ::-1]

plt.imshow(bgr_img)
plt.title("BGR Image (Reversed RGB)")
plt.axis('off')
plt.show()

# Step 4: Convert to Grayscale using simple formula
# Grayscale = 0.299*R + 0.587*G + 0.114*B
grayscale = 0.299 * img[..., 0] + 0.587 * img[..., 1] + 0.114 * img[..., 2]
grayscale = grayscale.astype(np.uint8)

plt.imshow(grayscale, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

# Step 5: Convert image array to Pandas DataFrame (for each channel)
r_df = pd.DataFrame(img[..., 0])
g_df = pd.DataFrame(img[..., 1])
b_df = pd.DataFrame(img[..., 2])

print("Red Channel DataFrame:\n", r_df.head())
print("Green Channel DataFrame:\n", g_df.head())
print("Blue Channel DataFrame:\n", b_df.head())

# Step 6: Modify a color channel (e.g., make red max everywhere)
img_mod = img.copy()
img_mod[..., 0] = 1  # Set red channel to max

plt.imshow(img_mod)
plt.title("Modified Image (Red Max)")
plt.axis('off')
plt.show()
```
